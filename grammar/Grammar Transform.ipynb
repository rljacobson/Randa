```python
file_name = "Miranda_expanded.ebnf"
f = open(file_name, 'w')
symbols = [] # Retains the ordering
definitions = dict()
for line in txt2.splitlines():
    [symbol, _, definition] = line.partition(':')
    if symbol in definitions:
        definitions[symbol].append(definition)
    else:
        symbols.append(symbol)
        definitions[symbol] = [definition]

for symbol in symbols:
    indent = (len(symbol)+1)*' '
    defs = definitions[symbol]
    
    end='\n'
    if len(defs) < 2:
        end=''
    
    if len(defs[0].strip()) == 0:
        print(f"{symbol} = epsilon", file=f, end=end)
    else:
        print(f"{symbol} = {defs[0]}", file=f, end=end)
    for d in defs[1:]:
        print(f"{indent}| {d}", file=f)
    
    if len(defs) < 2:
        print(f" ;\n", file=f)
    else:
        print(f"{indent};\n", file=f)
  
f.close()
```
